```python
import pandas as pd

sia205 = pd.read_csv("SIA205.20260226T150209.csv")
```

```python
sia205.head()
```

```python
deciles_2024 = sia205[
    (sia205["Year"] == 2024) &
    (sia205["Statistic Label"] == "Average equivalised nominal disposable income")
].copy()

deciles_2024 = deciles_2024[["Deciles", "VALUE"]]

deciles_2024["Monthly Income"] = deciles_2024["VALUE"] / 12

deciles_2024
```

```python
deciles_2024["Savings Capacity (15%)"] = deciles_2024["Monthly Income"] * 0.15
deciles_2024
```

```python
pregnancy_cost_model = {
    "Baby Equipment": 2500,
    "Clothing": 500,
    "Supplements": 200,
    "Classes": 200,
    "Travel & Misc": 600
}

total_cost = sum(pregnancy_cost_model.values())
total_cost
```

```python
months_remaining = 8
monthly_required = total_cost / months_remaining
monthly_required
```

```python
deciles_2024["Savings Capacity (15%)"]
```

```python
deciles_2024["Can Afford?"] = deciles_2024["Savings Capacity (15%)"] >= monthly_required
deciles_2024
```

```python
deciles_2024["Savings Capacity (20%)"] = deciles_2024["Monthly Income"] * 0.20
deciles_2024["Can Afford at 20%?"] = deciles_2024["Savings Capacity (20%)"] >= monthly_required
deciles_2024
```

**COST SCENARIOS**

```python
cost_scenarios = {
    "Essential": 3000,
    "Standard": 4500,
    "Premium": 6500
}

cost_scenarios
```

```python
months_remaining = 8

results = []

for scenario, cost in cost_scenarios.items():
    monthly_required = cost / months_remaining

    for index, row in deciles_2024.iterrows():
        results.append({
            "Scenario": scenario,
            "Decile": row["Deciles"],
            "Monthly Required": monthly_required,
            "Savings Capacity (15%)": row["Savings Capacity (15%)"],
            "Feasible": row["Savings Capacity (15%)"] >= monthly_required
        })

import pandas as pd
results_df = pd.DataFrame(results)

results_df.head(29)
```
